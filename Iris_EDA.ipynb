### Data Wrangling

```python
df.drop(columns=['petallenBin','petalwidBin','sepallenBin','sepalwidBin','index'],inplace=True)
```

```python
df.isnull().sum()
```

#### Outliers Treatment

```python
sns.boxplot(x='sepal_length',data=df)
```

```python
sns.boxplot(x='sepal_width',data=df)
```

```python
IQR = stats.iqr(df['sepal_width'])
```

```python
IQR
```

```python
Upper=df['sepal_width'].quantile(0.75)+1.5*IQR
Lower=df['sepal_width'].quantile(0.25)-1.5*IQR
print(Upper)
print(Lower)
```

```python
print(df['sepal_width'].quantile(0.95))
```

```python
upperouts = df[df['sepal_width']>Upper]
upperouts
```

```python
for i in upperouts.index:
    df.at[i,'sepal_width']=df['sepal_width'].quantile(0.95)
```

```python
lowerouts = df[df['sepal_width']<Lower]
lowerouts
```

```python
for i in lowerouts.index:
    df.at[i,'sepal_width']=df['sepal_width'].quantile(0.05)
```
